```python
df = sns.load_dataset('car_crashes')
print(df.head())
```

```python
plt.scatter(df.speeding,df.alcohol)
sns.set()
plt.show()
```

```python
plt.scatter(df.speeding,df.alcohol)
sns.set_style("whitegrid")
plt.show()
```

```python
plt.scatter(df.speeding,df.alcohol)
sns.set_style("ticks")
sns.despine()
plt.show()
```

```python
sns.palplot(sns.color_palette("deep", 10))
```

```python
tips = sns.load_dataset("tips")
sns.relplot(data=tips, x="total_bill", y="tip")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", col="day", col_wrap=2)
```

```python
sns.relplot(data=tips, x="size", y="tip",kind="line",errorbar=None)
```

```python
df = sns.load_dataset('iris')
sns.displot(df['petal_length'])
```

```python
sns.set_context('paper')
titanic = sns.load_dataset('titanic')
sns.barplot(x = 'embark_town', y = 'age', data = titanic,palette = 'PuRd',errorbar=None)
plt.legend()
plt.show()
print(titanic.columns)
```
